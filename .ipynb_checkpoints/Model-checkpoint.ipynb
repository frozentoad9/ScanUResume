```python
#importing the cleaned data
import pandas as pd
import matplotlib.pyplot as plt
resume_df = pd.read_csv('CleanedData.csv', encoding='utf-8')
resume_df = resume_df.drop('Unnamed: 0', axis=1)
print(resume_df.head())
print(resume_df.tail())
```

```python
#!pip install wordcloud
import nltk
from nltk.corpus import stopwords
import string
from wordcloud import WordCloud

stopwords = set(stopwords.words('english'))
vocab = []
sent_collec = ""
sentences = resume_df['cleaned_resume'].values
for i in range(len(sentences)):
    sent_collec += sentences[i]      #adding all sentences line by line
    word_token = nltk.word_tokenize(sentences[i])
    for word in word_token:
        if word not in stopwords and word not in string.punctuation:
            vocab.append(word)

#frequency of all the words
word_freq = nltk.FreqDist(vocab)
most_common = word_freq.most_common(50)
print(most_common)
```

```python
#creating a word cloud of the skills
wc = WordCloud().generate(sent_collec)
plt.figure(figsize = (15, 15))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
#lable encoding the labels (i.e. the categories in the dataset)
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
var_mod = ['Category']
for i in var_mod:
    resume_df[i] = le.fit_transform(resume_df[i])
```

```python
# preprocessing the data
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.sparse import hstack

text = resume_df.cleaned_resume.values
labels = resume_df.Category.values

word_vectorizer = TfidfVectorizer(max_features = 1500)
word_vectorizer.fit(text)
word_features = word_vectorizer.transform(text)

print("Word Features: {}".format(word_features.shape))

X_train, X_test, y_train, y_test = train_test_split(word_features, labels, random_state = 0, test_size=0.2)
print(X_train.shape)
print(X_test.shape)
```

```python
#modeling the data
from sklearn.svm import SVC
from sklearn.multiclass import OneVsRestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
# SVM
clf1 = SVC()
clf1.fit(X_train, y_train)
pred = clf1.predict(X_test)
print('accuracy on train data : {}'.format(clf1.score(X_train, y_train)))
print('accuracy on test data : {}'.format(clf1.score(X_test, y_test)))
```

```python
# GaussianNB
clf2 = OneVsRestClassifier(KNeighborsClassifier())
clf2.fit(X_train, y_train)
print('accuracy on train data : {}'.format(clf2.score(X_train, y_train)))
print('accuracy on test data : {}'.format(clf2.score(X_test, y_test)))
```

```python
# SVM gave the best accuracy
print('Classification report for classfier (SVC): ')
print(classification_report(y_test, pred))
```
